Una vez que se exploran los nombres de las columnas, los valores nulos y duplicados se procede a realizar las transformaciones:

1) Generar campo id: Cada id se compondrá de la primera letra del nombre de la plataforma, seguido del show_id ya presente en los datasets (ejemplo para títulos de Amazon = as123)

```python
# Crear la columna id:
amazon.insert(0, 'id', 'a' + amazon['show_id'].astype(str))

# Se crea la columna 'platform' ya que será usada para las consultas más adelante
amazon['platform']= 'Amazon'

# Se borra la columna 'show_id' ya que aporta información repetida
amazon = amazon.drop('show_id', axis=1) 

amazon.head()
```

```python
# Crear la columna id:
disney_plus.insert(0, 'id', 'd' + disney_plus['show_id'].astype(str))

# Se crea la columna 'platform' ya que será usada para las consultas más adelante
disney_plus['platform']= 'Disney Plus'

# Se borra la columna 'show_id' ya que aporta información repetida
disney_plus = disney_plus.drop('show_id', axis=1) 

disney_plus.head()
```

```python
# Crear la columna id:
hulu.insert(0, 'id', 'h' + hulu['show_id'].astype(str))

# Se crea la columna 'platform' ya que será usada para las consultas más adelante
hulu['platform']= 'Hulu'

# Se borra la columna 'show_id' ya que aporta información repetida
hulu = hulu.drop('show_id', axis=1) 

hulu.head()
```

```python
# Crear la columna id:
netflix.insert(0, 'id', 'n' + netflix['show_id'].astype(str))

# Se crea la columna 'platform' ya que será usada para las consultas más adelante
netflix['platform']= 'Netflix'

# Se borra la columna 'show_id' ya que aporta información repetida
netflix = netflix.drop('show_id', axis=1) 

netflix.head()
```

2) Los valores nulos del campo rating deberán reemplazarse por el string “G” (corresponde al maturity rating: “general for all audiences”:
3) De haber fechas, deberán tener el formato AAAA-mm-dd:
4) Los campos de texto deberán estar en minúsculas, sin excepciones:



```python
# Se reemplazan los valores nulos en 'rating' por 'G'
amazon['rating'] = amazon['rating'].fillna('G')

# Se cambia el formato de la fecha
amazon['date_added'] = pd.to_datetime(amazon['date_added'], errors='coerce').dt.strftime('%Y-%m-%d')

# Se cambian todos los campos de textos a minúsculas
amazon['type'] = amazon['type'].str.lower()
amazon['title'] = amazon['title'].str.lower()
amazon['director'] = amazon['director'].str.lower()
amazon['country'] = amazon['country'].str.lower()
amazon['rating'] = amazon['rating'].str.lower()
amazon['duration'] = amazon['duration'].str.lower()
amazon['listed_in'] = amazon['listed_in'].str.lower()
amazon['description'] = amazon['description'].str.lower()
amazon['platform'] = amazon['platform'].str.lower()

amazon.head()
```

```python
nuevo_amazon = amazon
```

```python
nuevo_amazon = nuevo_amazon.assign(duration_int=nuevo_amazon['duration'].str.extract(r'(\d+)', expand=False).fillna(0).astype(int),
                        duration_type=nuevo_amazon['duration'].str.extract(r'(min|season|seasons)', expand=False))

nuevo_amazon.head()
```

```python
nuevo_disney_plus = disney_plus
```

```python
nuevo_disney_plus = nuevo_disney_plus.assign(duration_int=nuevo_disney_plus['duration'].str.extract(r'(\d+)', expand=False).fillna(0).astype(int),
                   duration_type=nuevo_disney_plus['duration'].str.extract(r'(min|season|seasons)', expand=False))

nuevo_disney_plus.head()
```

```python
nuevo_hulu = nuevo_hulu.assign(duration_int=nuevo_hulu['duration'].str.extract(r'(\d+)', expand=False).fillna(0).astype(int),
                   duration_type=nuevo_hulu['duration'].str.extract(r'(min|season|seasons)', expand=False))

nuevo_hulu.head()
```

```python
nuevo_netflix = nuevo_netflix.assign(duration_int=nuevo_netflix['duration'].str.extract(r'(\d+)', expand=False).fillna(0).astype(int),
                   duration_type=nuevo_netflix['duration'].str.extract(r'(min|season|seasons)', expand=False))

nuevo_netflix.head()
```

Unir los cuatro dataframes para ver la totalidad de datos faltantes:

```python
combined = pd.concat([nuevo_amazon, nuevo_disney_plus, nuevo_hulu, nuevo_netflix])
combined.to_csv('streaming.csv', index=False)
```
